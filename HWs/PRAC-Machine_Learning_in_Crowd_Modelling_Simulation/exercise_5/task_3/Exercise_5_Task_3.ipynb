```python
from linear_solver import LinearSolver
from nonlinear_solver import NonlinearSolver
from trajectory import Trajectory
```

```python
solver = LinearSolver(file_name_x0="nonlinear_vectorfield_data_x0.txt", file_name_x1="nonlinear_vectorfield_data_x1.txt", folder_path='input')
solver.solve_linear()
```

```python
L_and_eps_list = [(250, .1), (500, .25), (750, .1), (1000, .1)]
for L_and_eps in L_and_eps_list:
    L, eps = L_and_eps
    solver = NonlinearSolver(L=L, eps=eps, file_name_x0="nonlinear_vectorfield_data_x0.txt", file_name_x1="nonlinear_vectorfield_data_x1.txt", folder_path='input')
    x_0, x_l, coef = solver.solve_nonlinear()
```

```python
L, eps = (1000, .1)
solver = NonlinearSolver(L=L, eps=eps, file_name_x0="nonlinear_vectorfield_data_x0.txt", file_name_x1="nonlinear_vectorfield_data_x1.txt", folder_path='input')
x_0, x_l, coef = solver.solve_nonlinear()
```

```python
trajectory = Trajectory(x_0, x_l, coef, L, eps)
close_points = trajectory.plot_trajectory()
```

```python
for close_point in close_points:
    print("Previous point: {}, Next point: {}, Distance: {}".format(close_point[0], close_point[1], close_point[2]))
```
