```python
import numpy as np
import matplotlib as plt
%matplotlib inline
```

```python
def logistic_map(x, r):
    """
    Representation of the logistic map
    """
    return r * x * (1 - x)
```

```python
def draw_bifurcation_diagram(r_range):
    """
    Draws the bifurcation diagram of logistic map for the given r range
    """
    eps = 0.01; x = eps
    r = r_range[0]; dr = 0.005

    seen_r = []; seen_x = []

    # r in [0, 4]
    while r <= r_range[1]:
        # Do nothing for the first 500 iterations to make x converge to the equilibrium
        for _ in range(500): x = logistic_map(x, r)
        for _ in range(1000):
            x = logistic_map(x, r)
            seen_r.append(r)
            seen_x.append(x)
        r += dr
        x += eps

    from matplotlib.pyplot import figure

    figure(figsize=(12, 8), dpi=80)
    plt.pyplot.scatter(seen_r, seen_x, s=1, linewidths=0.001, c='black')
    plt.pyplot.xlabel('r')
    plt.pyplot.ylabel('dx/dt')
    plt.pyplot.title('Bifurcation diagram for Logistic map')
    plt.pyplot.axis([0.0, 4.0, -1.0, 1.0])
    plt.pyplot.show()
```

```python
def run():
    """
    Calls the function to draw bifurcation diagram with corresponding r range.
    """
    selection = int(input('For which interval should the diagram be drawn?\n1- (0, 2]\n2- [2, 4]\n3- (0, 4]\n'))
    r_ranges = [[0, 2], [2, 4], [0, 4]]
    draw_bifurcation_diagram(r_ranges[selection-1])
```

```python
# Run this cell to draw bifurcation diagrams

run()
```
