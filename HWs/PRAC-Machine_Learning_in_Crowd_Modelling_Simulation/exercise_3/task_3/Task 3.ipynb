```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import interactive
interactive(True)
%matplotlib notebook
# interactive is imported for better observation of the 3D surface. And other libraries are imported for better usability


# x and alpha 2 are sampled just like recommended in exercise sheet 
x = np.linspace(-5, 5, 40)
alp2 = np.linspace(-5,5,40)
temp = x




# X and Y are transform into the meshgrid structure 
X, Y = np.meshgrid(alp2,x)

print(np.shape(X))
print(np.shape(Y))
# the alpha 1 value is calculated from the equation for the each meshgrid position
Z = -X*Y + Y**3 


fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
#surf = ax.plot_surface(X,Y,Z)

ax.set_xlabel('alpha1')
ax.set_ylabel('alpha2')
ax.set_zlabel('x')
ax.set_xlim(-20, 20)
ax.set_ylim(-20,20)
ax.set_zlim(-20,20)
#surface is plotted with X,Y,Z grids
ax.plot_surface(Z, X, Y, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none');

```
